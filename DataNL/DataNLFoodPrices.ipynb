```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```
url = 'https://raw.githubusercontent.com/Saroramath/MachineLearning/main/DataExamples/1810024501_databaseLoadingData.csv'
df = pd.read_csv(url, error_bad_lines=False)
```

```
df
```

```
print(df.columns.tolist())
```

```
df['Products'].unique()
```

```
df_banana=df.loc[lambda df: df['Products'] == 'Bananas, per kilogram', :]
```

```
df_banana.info()
```

```
df_banana['VALUE'].mean()
```

```
date  = df_banana['REF_DATE']
```

```
plt.figure(figsize=(30,10))
plt.plot(date,df_banana['VALUE'])
plt.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    labelbottom=False) 
plt.xlabel("2007-2020",fontsize=25)
plt.ylabel("Price",fontsize=25)
```

```
df_orange=df.loc[lambda df: df['Products'] == 'Oranges, per kilogram', :]
```

```
df_orange.info()
```

```
df_ob = [df_banana,df_orange]
```

```
df_ob = pd.merge(df_banana, df_orange, on="REF_DATE")
```

```
df_ob.info()
```

```
plt.figure(figsize=(30,10))
plt.plot(date, df_banana['VALUE'])
plt.plot(date, df_orange['VALUE'])
plt.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    labelbottom=False) 
plt.xlabel("2007-2020",fontsize=25)
plt.ylabel("Price",fontsize=25)
```

```
import plotly.graph_objects as go

# Create random data with numpy
import numpy as np
np.random.seed(1)

config= {'displaylogo': False}
# Create traces
fig = go.Figure()
fig.add_trace(go.Scatter(x=df2_banana.REF_DATE, y=df2_banana.VALUE,
                    mode='lines+markers',
                    name='PEI'))
fig.add_trace(go.Scatter(x=df_banana.REF_DATE, y=df_banana.VALUE,
                    mode='lines+markers',
                    name='NL'))

fig.update_layout(title = 'Bananas',yaxis_title = 'Price', yaxis_showgrid = False, xaxis_showgrid = False, plot_bgcolor = 'white')
fig.update_layout(hovermode="x")

fig.show(config= {'displaylogo': False})

```

```
fig.write_html("file.html", include_plotlyjs='cdn',config= {'displayModeBar': False} )
```

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```
url = 'https://raw.githubusercontent.com/Saroramath/MachineLearning/main/DataNLD/3210004701_databaseLoadingData-2.csv'
df = pd.read_csv(url, error_bad_lines=False)
```

```
df
```

```
print(df.columns.tolist())
```

```
df['GEO'].unique()
```

```
df_Canada=df.loc[lambda df: df['GEO'] == 'Canada', :]
```

```
df_Canada.info()
```

```
df_Canada['VALUE'].mean()
```

```
date  = df_Canada['REF_DATE']
```

```
plt.figure(figsize=(30,10))
plt.plot(date,df_Canada['VALUE'])
plt.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    labelbottom=False) 
plt.xlabel("2007-2020",fontsize=25)
plt.ylabel("Price",fontsize=25)
```

```
df_NL=df.loc[lambda df: df['GEO'] == 'Newfoundland and Labrador', :]
```

```
df_NL.info()
```

```
plt.figure(figsize=(30,10))
#plt.plot(date, df_Canada['VALUE'])
plt.plot( df_NL['VALUE'])
plt.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    labelbottom=False) 
plt.xlabel("2007-2020",fontsize=25)
plt.ylabel("Price",fontsize=25)
```

## Plotlly plots

```
import plotly.graph_objects as go

# Create random data with numpy
import numpy as np
np.random.seed(1)

config= {'displaylogo': False}
# Create traces
fig = go.Figure()
fig.add_trace(go.Scatter(x=df_Canada.REF_DATE, y=df_Canada.VALUE,
                    mode='lines+markers',
                    name='Canada'))
fig.add_trace(go.Scatter(x=df_PIE.REF_DATE, y=df_PIE.VALUE,
                    mode='lines+markers',
                    name='PIE'))
fig.add_trace(go.Scatter(x=df_Ont.REF_DATE, y=df_Ont.VALUE,
                    mode='lines+markers',
                    name='Ontario'))
fig.add_trace(go.Scatter(x=df_Qc.REF_DATE, y=df_Qc.VALUE,
                    mode='lines+markers',
                    name='Quebec'))
fig.add_trace(go.Scatter(x=df_NB.REF_DATE, y=df_NB.VALUE,
                    mode='lines+markers',
                    name='NB'))
fig.add_trace(go.Scatter(x=df_Ab.REF_DATE, y=df_Ab.VALUE,
                    mode='lines+markers',
                    name='Alberta'))
fig.add_trace(go.Scatter(x=df_BC.REF_DATE, y=df_BC.VALUE,
                    mode='lines+markers',
                    name='BC'))
fig.add_trace(go.Scatter(x=df_NL.REF_DATE, y=df_NL.VALUE,
                    mode='lines+markers',
                    name='NL'))


fig.update_layout(title = 'Value per acre of farm land and buildings',yaxis_title = 'Price', yaxis_showgrid = False, xaxis_showgrid = False, plot_bgcolor = 'white')
fig.update_layout(hovermode="x")

fig.show(config= {'displaylogo': False})

```

```
fig.write_html("file.html", include_plotlyjs='cdn',config= {'displayModeBar': False} )
```
